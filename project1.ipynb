```python
#string==> loation, fuel_type, #transmission , owner_type

#get_dummy

dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
#combain two data frame
dfe=pd.concat([df,dummy],axis=1)
dfe
```

```python
dfe=dfe.drop(['Unnamed: 0','Location','Fuel_Type','Power','Transmission','Owner_Type','Name','New_Price'],axis=1)
dfe
```

```python
print(dfe.isna().sum())
```

```python
#string replace(eg:milage 22 kmpl==>22)
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
#null
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe
```

```python
dfe.dtypes
```

```python
#covert object to int or flot
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Mileage']=dfe['Mileage'].astype(float)
```

```python
dfe.dtypes
```

```python
dfe.isna().sum()
```

```python
#replace zero value (mileage,power,engine)
import numpy as np
dfe.loc[dfe.Engine==0,'Engine']=np.NaN #not a number  #read as missing value
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
```

```python
dfe.isna().sum()
```

```python
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
```

```python
dfe.isna().sum()
```

```python
x=dfe['Price']
x
```

```python
y=dfe.drop(['Price'],axis=1)
```
